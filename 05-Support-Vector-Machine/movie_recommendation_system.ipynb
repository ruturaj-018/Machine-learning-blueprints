scikit-learn's __CountVectorizer()__ is used to convert text documents into a vector

cosine_similarity__ of the sci-kit learn is used for the metric of similarity matrix

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv("movie_dataset.csv")
```

```python
df.head(5)
```

## Features list
We'll choose the features which are most relevant to us i.e. _*keywords, cast, genres, director*_

```python
features = ['keywords', 'cast', 'genres', 'vote_average']
```

### removing null values

```python
for feature in features:
  df[feature] = df[feature].fillna('')
```

## Combined Features List
combining the relevant features into a single feature

```python
def combined_features(row):
    return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']

df['combined_features'] = df.apply(combined_features,axis = 1)
```

## Extracting Features
1. __sklearn.feature_extraction__ module can be used to extract features in format supported by machine learning algorithms
2. __CountVectorizer()'s__ _*fit_transform*_ will help to count the number of text

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df['combined_features'])
print("Count Matrix: ",count_matrix.toarray())
```

## Cosine Similarity
We'll use the cosine_similarity as a metric to compute similarity between two items.

__cosine_sim__ is a numpy array with calculated cosine-similarity between two movies.

```python
cosine_sim = cosine_similarity(count_matrix)
```


now we'll take the input movies that user like in the __movie_user_like__ variable.

we need to know the content user like in order to predict the similar.

```python
movie_user_like = "Dead Poets Society"

def get_index_from(title):
    return df[df.title == title]["index"].values[0]

movie_index = get_index_from(movie_user_like)
```

## Generating similar movies matrix

```python
similar_movies = list(enumerate(cosine_sim[movie_index]))
```

```python
print(similar_movies)
```

## Sorting the similar movies in descending order

```python
sorted_similar_movies = sorted(similar_movies, key = lambda x:x[1], reverse = True)
```

```python
print(sorted_similar_movies)
```

## Printing the  Similar Movies

```python
def get_title_from_index(index):
    return df[df.index == index]["title"].values[0]

i=0
for movies in sorted_similar_movies:
    print(get_title_from_index(movies[0]))
    i = i+1;
    if i>15:
        break
```
